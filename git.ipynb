### Установка Cookiecutter

Cookiecutter — это утилита для быстрого создания новых проектов на основе шаблонов. Чтобы установить Cookiecutter, выполните следующие шаги:

#### Для Windows:
```bash
pip install --user cookiecutter
```

После установки добавьте путь к исполняемым файлам Python в переменную среды `PATH`. Обычно это каталог вида:
```
C:\Users\<ИмяПользователя>\AppData\Roaming\Python\Python38\Scripts
```

#### Для macOS и Linux:
```bash
pip3 install --user cookiecutter
```

#### Запуск Cookiecutter

После правильной настройки пути можно запустить Cookiecutter простым вводом команды:
```bash
cookiecutter
```

Если возникают проблемы с поиском команды, запустите её через модуль Python:
```bash
python -m cookiecutter # в Windows
python3 -m cookiecutter # в macOS или Linux
```

При работе с Cookiecutter рекомендуется нажимать клавишу Enter для принятия значений по умолчанию, если это подходит для вашего случая.

##### Как Git отслеживает статус файлов
Для репозитория Git неотслеживаемые файлы в рабочей копии не существуют.

Три (3) состояния отслеживаемых файлов:
В сохраненном (закрепленном) состоянии файл в рабочей копии идентичен
последнему коммиту в репозитории. (Иногда это состояние называется не-
измененным, или чистым.)


##### Основные этапы работы с Git:

Создание/Клонирование репозитория: git init или git clone.
Добавление файлов для отслеживания: git add <имя_файла>.
Сохранение изменений (коммит): git commit -am "<сообщение коммита>".
Для получения справки по команде используйте: git help <команда>.


##### Состояния файлов в Git:

Отслеживаемые файлы: Добавлены и сохранены в репозитории. Имеют три состояния:

Сохраненный (закрепленный): Файл идентичен последнему коммиту.
Измененный: Файл отличается от последнего коммита.
Индексированный (подготовленный): Файл изменен и помечен для включения в следующий коммит.
Неотслеживаемые файлы: Не добавлены в репозиторий. Git их игнорирует.

Переходы между состояниями:

git add <файл>: Переводит неотслеживаемый файл в отслеживаемый и индексированный.
Изменение сохраненного файла автоматически переводит его в измененное состояние.
git status: Показывает текущий статус репозитория и состояние файлов.


##### Назначение индексирования в Git:

Индексирование (staging) позволяет подготовить определенные изменения в файле к коммиту, не включая все изменения сразу. В индексе хранятся описания изменений, а не полные файлы, что позволяет коммитить частичные изменения.

Сложности индексирования:

Область индексирования может быть сложной из-за особых случаев (например, файл может быть изменен после индексации). Это приводит к сложностям и потенциальной дезинформации в источниках о Git.

Рекомендации автора по упрощению работы:

Использовать git commit -am для одновременного индексирования и коммита.
Сохранять файлы сразу после добавления, переименования или удаления.
Использовать графические инструменты Git для упрощения работы.

##### Создание Git-репозитория локально:

Git - распределенная система контроля версий, хранит коммиты и метаданные в папке .git локально.
Для создания репозитория используется команда git init в нужной папке. (В macOS и Linux - mkdir для создания папки).
git init создает папку .git с метаданными, превращая папку в репозиторий.
Файлы в папке .git изменять не следует.
Локальный репозиторий - на вашем компьютере, удаленный - на другом.
После создания репозитория:

Можно использовать git для добавления файлов и отслеживания изменений.
git status покажет статус репозитория. Сразу после git init покажет, что коммитов нет.

##### Добавление файлов для отслеживания (git add):

Только отслеживаемые файлы можно коммитить и откатывать.
git status показывает статус файлов. Неотслеживаемые файлы отображаются в секции “Untracked files”.
git add переводит файлы из неотслеживаемого или измененного состояния в индексированное.
Способы использования git add:

git add <файл>: Добавляет конкретный файл.
git add *.py: Добавляет все файлы .py в текущей директории и поддиректориях.
git add .: Добавляет все неотслеживаемые файлы в текущей директории.
После выполнения git add:

git status покажет проиндексированные файлы в секции “Changes to be committed”.
Статус “new file” означает, что файлы новые и были добавлены в репозиторий.

##### Игнорирование файлов в Git (.gitignore):

Неотслеживаемые файлы отображаются в git status. Некоторые файлы (временные, скомпилированные, конфиденциальные данные) следует исключить из отслеживания.
Для этого создается файл .gitignore в корне репозитория. В нем перечисляются файлы и папки, которые Git должен игнорировать.
Git автоматически исключает их из git add, git commit и они не отображаются в git status.
Файл .gitignore также добавляется в репозиторий для других разработчиков.
Содержимое файла .gitignore:

Шаблоны используют символ *.
Комментарии начинаются с #.
Примеры игнорируемых файлов:

Временные файлы
.pyc, .pyo, .pyd (скомпилированные Python-файлы)
Папки .tox, htmlcov (генерируемые инструменты разработчика)
Скомпилированные или сгенерированные файлы
Файлы с паролями, токенами, номерами кредитных карт
Просмотр игнорируемых файлов:

git ls-files --other --ignored --exclude-standard

Сохранение изменений (коммит):

Для сохранения всех измененных и проиндексированных файлов:
git commit -am "<сообщение>" (индексирует и коммитит).
Для сохранения только выбранных файлов:
git commit -m "<сообщение>" <файл1> <файл2> (не использует -a, указывает файлы).
Важность сообщений коммитов:

Сообщения должны быть информативными (например, “Fixed the currency conversion bug.”).
Избегайте общих сообщений (“Обновленный код”, “Исправлены ошибки”).
Качественные сообщения важны, особенно в больших командных проектах.
Работа с редактором Vim (при отсутствии -m):

Если забыли -m, Git откроет Vim.
Для выхода из Vim без сохранения: Esc, затем qa!.
Другие команды:

git status после коммита показывает, что рабочее дерево чистое.
Git не сохраняет пустые папки; папки включаются автоматически при сохранении файлов в них.
Переписать сообщение последнего коммита: git commit --amend -m "<новое_сообщение>"


##### Просмотр изменений перед коммитом (git diff):

Команда git diff показывает различия между рабочей копией файла и последним сохраненным состоянием (коммитом).
Строки, начинающиеся с -, были удалены.
Строки, начинающиеся с +, были добавлены.
Пример использования:

Создать или открыть файл README.md.
Внести в него изменения.
Выполнить git diff для просмотра этих изменений.
После просмотра изменений и их исправления, их можно добавить (git add) и сохранить (git commit).


##### Просмотр изменений в графическом интерфейсе (git difftool):

Для более удобного просмотра изменений используйте графические инструменты.
Примеры инструментов:
Windows: WinMerge (https://winmerge.org/)
Linux: Meld ( sudo apt-get install meld) или Kompare ( sudo apt-get install kompare)
macOS: tkdiff (установка через Homebrew: /bin/bash -c "$(curl -fsSL https://raw.githubusercontent.com/Homebrew/install/master/install.sh)" и brew install tkdiff)
Настройка Git для использования инструментов:
git config diff.tool <название>, где <название> - winmerge, tkdiff, meld или kompare.
git difftool <имя_файла> для запуска выбранного инструмента.
Отключение запроса подтверждения:
git config --global difftool.prompt false
Можно настроить эти инструменты (или использовать встроенные средства сравнения) в графических Git-клиентах.

##### Частота сохранения изменений (коммитов):

Цель: Найти баланс между детализацией истории и удобством навигации.
Слишком часто: Множество мелких коммитов, затрудняющих поиск нужной версии.
Слишком редко: Большие коммиты, откат к которым может отменить слишком много изменений.
Рекомендации по частоте коммитов:

Сохранять при:
Завершении блока функциональности.
Завершении класса.
Исправлении ошибки.
Не сохранять:
Код с синтаксическими ошибками.
Очевидно неработоспособный код.
Обязательно:
Выполнять все модульные тесты перед сохранением.
В идеале, все тесты должны проходить. Если тесты не проходят, упомянуть об этом в сообщении коммита.
Важно: Коммиты могут содержать от нескольких строк до сотен строк измененного кода, но всегда должны представлять собой рабочую версию программы.

##### Удаление файлов из репозитория (git rm):

Важно: Не удалять файлы командами del (Windows) или rm (macOS/Linux) напрямую!
Используйте git rm <имя_файла>, чтобы удалить файл из рабочей копии и перестать отслеживать его.
git rm индексирует удаление, как и git add индексирует добавление.
Процесс удаления:
git rm <имя_файла>: Удаляет файл из рабочей копии и индексирует удаление.
git commit -m "<сообщение>": Фиксирует удаление в репозитории.

Важные моменты:
Удаленный файл все еще существует в истории репозитория.
git rm работает только с файлами в чистом (сохраненном) состоянии. Если есть несохраненные изменения, Git предложит сохранить или отменить их. Можно отменить изменения командой git reset HEAD <имя_файла>.
Для восстановления файла, если он был удалён ошибочно, можно использовать команду git restore <имя_файла>.

##### Переименование и перемещение файлов в репозитории (git mv):

Не переименовывайте или перемещайте файлы напрямую в файловой системе (например, с помощью mv или через проводник/файловый менеджер).
Используйте команду git mv <старое_имя> <новое_имя>, а затем git commit.
git mv учитывает историю изменений файла при переименовании.
Процесс переименования/перемещения:
git mv <старое_имя> <новое_имя>: Переименовывает/перемещает файл и индексирует изменение.
git commit -m "<сообщение>": Фиксирует переименование/перемещение.

Перемещение в новую папку: Создайте папку (например, movetest).
git mv README.txt movetest/README.txt
git commit -m "<сообщение>"

##### Просмотр журнала коммитов (git log):

git log: Выводит список всех коммитов в репозитории.
Для навигации по журналу используйте клавиши ↑ и ↓. Нажмите q для выхода.
Каждый коммит имеет уникальный хеш (40 шестнадцатеричных символов).
В большинстве случаев достаточно первых 7 символов хеша.
Опции git log:
git log --oneline: Отображает сокращенный хеш и первую строку сообщения коммита.
git log --oneline -n <количество>: Отображает последние <количество> коммитов в сокращенном формате.

Просмотр содержимого файла в конкретном коммите:
git show <хеш>:<имя_файла>: Показывает содержимое файла <имя_файла> в коммите с хешем <хеш>.

##### Восстановление старых изменений:

Системы контроля версий только добавляют информацию; удаленные файлы можно восстановить.
Отмена изменения - это добавление нового изменения, которое возвращает состояние к предыдущему коммиту.
Конкретная команда для восстановления зависит от текущего состояния рабочей копии и индекса.
Ссылка на статью о различных видах отмены изменений:

https://github.blog/2015-06-08-how-to-undo-almost-anything-with-git/ (Предоставляет детальную информацию о способах отмены изменений в Git).
Важно: Прежде чем восстанавливать, нужно учитывать текущее состояние файлов.

##### Отмена несохраненных локальных изменений (git restore):

Используется, когда вы изменили файл в рабочей копии, но ещё не добавили его в индекс (git add) и не сделали коммит (git commit).
git restore <имя_файла>: Возвращает файл к состоянию последнего коммита.
По сути, это отмена изменений, внесенных в файл, которые не были зафиксированы.
Важно: Эту операцию отмены нельзя отменить.
git checkout .: Отменяет все изменения во всех файлах рабочей копии (альтернативный способ).

##### Деиндексирование проиндексированного файла (git restore –staged):

Используется, когда вы уже добавили файл в индекс (git add), но хотите исключить его из следующего коммита.
git restore --staged <имя_файла>: Возвращает файл из индексированного состояния обратно в состояние “измененный, но не проиндексированный”.
При выполнении git status он будет отображаться в разделе “Changes not staged for commit”.

##### Отмена нескольких последних коммитов (git revert -n):

Используется для отмены нескольких последних коммитов и возврата к более раннему состоянию.
git revert -n HEAD~<количество>..HEAD: Отменяет <количество> последних коммитов, но не создает коммит автоматически. Нужно сделать git add . и git commit чтобы зафиксировать изменения отмены.
После отмены изменений, необходимо добавить их в индекс (git add .) и зафиксировать (git commit -m "<сообщение>").
HEAD~<количество> означает “коммит, отстоящий на <количество> коммитов назад от HEAD”.
История коммитов не удаляется, отмененные коммиты остаются в истории.
git revert создает новые коммиты, которые отменяют изменения, внесенные предыдущими коммитами. Это отличается от git reset, который может изменять историю.
Если нужно отменить “отмену”, можно снова использовать git revert на коммите, отменяющем изменения. Это восстановит эффект отменённых коммитов.

#####  Возврат к конкретному коммиту для отдельного файла

Чтобы откатить изменения конкретного файла до состояния определенного коммита, не затрагивая остальные файлы проекта, выполните следующие шаги:

Проверка истории изменений

Для начала посмотрите историю изменений вашего репозитория:
```bash
git log --oneline
```
Просмотр содержимого файла на нужный момент

Вы можете посмотреть содержимое файла на определенный момент времени с помощью команды:
```bash
git show 009b7c0:eggs.py
```

Эта команда выведет содержание файла `eggs.py`, зафиксированное в указанном коммите (`009b7c0`).

Откат изменений файла

Теперь восстановите файл обратно до нужного состояния:
```bash
git checkout 009b7c0 -- eggs.py
```

Это вернет файл `eggs.py` к предыдущей версии, указанной в хеше коммита.

Затем сохраните изменения:
```bash
git add eggs.py
git commit -m "Вернул яйца к состоянию 009b7c0"
```
##### Перезапись истории коммитов

Проблема
Иногда возникает ситуация, когда в репозиторий попадает конфиденциальная информация (например, пароли, API-ключи). Просто удалить такие данные и совершить новый коммит недостаточно, поскольку каждый, имеющий доступ к репозиторию, сможет просмотреть предыдущие версии и увидеть секретные данные.

Решение
Удаление конфиденциальной информации из репозитория навсегда требует специальных инструментов. Существует два основных способа решения проблемы:

1. **Использование команды** `git filter-branch`
   Эта команда позволяет переписывать историю коммитов, удаляя нежелательные данные из каждого коммита.
   
2. **Программа BFG Repo-Cleaner**
   Это специализированный инструмент, предназначенный именно для удаления больших объектов и конфиденциальных данных из истории Git. Считается предпочтительнее `git filter-branch`, так как работает быстрее и проще.

Обе методики подробно изложены в руководстве GitHub: [Removing sensitive data from a repository](https://help.github.com/en/articles/removing-sensitive-data-from-a-repository).

---

#### Превентивные меры
Рекомендуемый способ избежать попадания конфиденциальной информации в репозиторий:
- Создать отдельный файл (например, `secrets.txt` или `confidential.py`) для хранения секретных данных.
- Включить этот файл в `.gitignore`, чтобы предотвратить его попадание в репозиторий.
- Программный код может считывать необходимые секреты из указанного файла, что является безопасным решением по сравнению с хранением чувствительной информации прямо в исходниках.

##### ### Использование GitHub и команда `git push`

GitHub — крупнейший онлайн-сервис для хостинга проектов, использующих систему управления версиями Git. Репозитории Git, размещённые на GitHub, делают ваш проект доступным другим разработчикам для совместной работы и улучшения.

#### Зачем нужен GitHub?
- Возможность хранить резервную копию вашего проекта в облаке.
- Обеспечение доступности проекта для коллег и сообщества разработчиков.
- Совместная работа возможна независимо от статуса локального компьютера.

#### Как опубликовать проект на GitHub?
Используйте команду `git push`, чтобы отправить ваш локальный репозиторий на сервер GitHub. Например:

```bash
git remote add origin https://github.com/yourusername/yourrepository.git
git branch -M main
git push -u origin main
```

Где:
- `origin`: название удалённого сервера (обычно используется по умолчанию).
- `yourusername`: ваше имя пользователя на GitHub.
- `yourrepository`: название репозитория на GitHub.

Эти команды позволят синхронизировать ваш локальный репозиторий с удалённым хранилищем на GitHub.

Помните различия между терминологией:
- **Git**: Система контроля версий, используемая для отслеживания изменений в проекте.
- **GitHub**: Онлайн-платформа для хостинга репозиториев Git, обеспечивающая удобный интерфейс для совместного использования и сотрудничества.

##### ## Отправка существующего репозитория на GitHub

Чтобы отправить существующий локальный репозиторий на GitHub, выполните следующие шаги:

### Шаг 1: Подключение удалённого репозитория
Подключите удалённый репозиторий на GitHub к вашему локальному проекту с помощью следующей команды:

```bash
git remote add origin https://github.com/<ваш_логин>/<название_репозитория>.git
```

Замените `<ваш_логин>` и `<название_репозитория>` соответствующими значениями.

### Шаг 2: Отправка изменений
Отправьте изменения из локальной ветки в удалённую ветку с помощью команды:

```bash
git push -u origin master
```

Система попросит вас ввести логин и пароль от аккаунта GitHub.

### Примечания
- Первый аргумент (`-u`) связывает вашу локальную ветвь с удалённой.
- После первой успешной отправки последующие обновления можно отправлять простой командой:

```bash
git push
```

Это означает, что ваша локальная ветка была отправлена на удалённый сервер, и теперь её можно просматривать на странице GitHub.

### Рекомендации
- Регулярная отправка обновлений гарантирует согласованность между вашей локальной копией и репозиторием на GitHub.
- Загрузка репозитория на GitHub позволяет делиться проектом с коллегами и сообществом, получать обратную связь и сотрудничать с другими пользователями.

##### ## Клонирование существующего репозитория GitHub

Клонирование репозитория с GitHub на локальный компьютер выполняется с помощью команды `git clone`. Этот процесс полезен, если:

- Вы создаёте новый проект на GitHub и хотите перенести его на свой компьютер.
- Ваш локальный репозиторий оказался повреждён, и вы решили начать заново.

### Процесс клонирования

1. **Создайте репозиторий на GitHub**. Для нового проекта убедитесь, что установлен флажок *Initialize this repository with a README*.

2. Скопируйте URL репозитория, нажав кнопку *Clone or download*. URL будет иметь вид:

   ```
   https://github.com/<ваш_логин>/<название_репозитория>.git
   ```

3. Откройте терминал и выполните следующую команду:

   ```bash
   git clone https://github.com/<ваш_логин>/<название_репозитория>.git
   ```


Преимущества клонирования

- Простота восстановления поврежденного локального репозитория.
- Быстрая настройка рабочего окружения на новом устройстве.
- Возможность продолжить работу над проектом с другого устройства.

##### Словарь терминов для Git и GitHub

Git или Гит — система контроля и управления версиями файлов.

GitHub или Гитхаб — веб-сервис для размещения репозиториев и совместной разработки проектов.

Репозиторий Git — каталог файловой системы, в котором находятся: файлы конфигурации, файлы журналов операций, выполняемых над репозиторием, индекс расположения файлов и хранилище, содержащее сами контролируемые файлы.

Локальный репозиторий — репозиторий, расположенный на локальном компьютере разработчика в каталоге. Именно в нём происходит разработка и фиксация изменений, которые отправляются на удалённый репозиторий.

Удалённый репозиторий — репозиторий, находящийся на удалённом сервере. Это общий репозиторий, в который приходят все изменения и из которого забираются все обновления.

Форк (Fork) — копия репозитория. Его также можно рассматривать как внешнюю ветку для текущего репозитория. Копия вашего открытого репозитория на Гитхабе может быть сделана любым пользователем, после чего он может прислать изменения в ваш репозиторий через пулреквест.


Обновиться из апстрима — обновить свою локальную версию форка до последней версии основного репозитория, от которого сделан форк.

Обновиться из ориджина — обновить свою локальную версию репозитория до последней удалённой версии этого репозитория.

Клонирование (Clone) — скачивание репозитория с удалённого сервера на локальный компьютер в определённый каталог для дальнейшей работы с этим каталогом как с репозиторием.

Ветка (Branch) — это параллельная версия репозитория. Она включена в этот репозиторий, но не влияет на главную версию, тем самым позволяя свободно работать в параллельной. Когда вы внесли нужные изменения, то вы можете объединить их с главной версией.

Мастер (Master) — главная или основная ветка репозитория.

Коммит (Commit) — фиксация изменений или запись изменений в репозиторий. Коммит происходит на локальной машине.

Пул (Pull) — получение последних изменений с удалённого сервера репозитория.

Пуш (Push) — отправка всех неотправленных коммитов на удалённый сервер репозитория.

Пулреквест (Pull Request) — запрос на слияние форка репозитория с основным репозиторием. Пулреквест может быть принят или отклонён вами, как владельцем репозитория.

Мёрдж (Merge) — слияние изменений из какой-либо ветки репозитория с любой веткой этого же репозитория. Чаще всего слияние изменений из ветки репозитория с основной веткой репозитория.

Кодревью — процесс проверки кода на соответствие определённым требованиям, задачам и внешнему виду.
